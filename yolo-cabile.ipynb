```python
%%writefile yolo_cabile.yaml
train: /kaggle/input/cabile-2/train/images
val: /kaggle/input/cabile-2/validation/images
names:  
  0: cabile
 
```

```python
import wandb
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("WANDB_API_KEY") 
wandb.login(key=secret_value_0)
```

```python
model = YOLO("yolov8n.pt")   
results = model.train(data="/kaggle/working/yolo_cabile.yaml", 
                      epochs=300, 
                      imgsz=1024,
                      half= True,
                      warmup_epochs= 10,
                      optimizer= "Adam",
                      cos_lr= True,
                      lr0=0.001,
                      lrf= 0.0001,
                      batch= -1
                     )
```
